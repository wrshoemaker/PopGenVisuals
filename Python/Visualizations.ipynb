```python
import ipywidgets as widgets
import popGenSims as pgs
import popGenVisuals as pgv
import  matplotlib.pyplot as plt
%matplotlib inline
```

```python
#widgets.interact(pgv.driftFig(mid_N = True),
#               N = (10, 100, 10))
def plott(s):
    sims = pgs.multipleDriftSims(N = 100, n1 = 1, s = s, reps = 50)
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)
    plt.xlabel('Generation', fontsize=20)
    plt.ylabel('Frequency of beneficial allele', fontsize=20)
    for sim in sims:    
        ax.plot(sim[0], sim[1], c = 'b')
    fig.canvas.draw()
widgets.interact(plott, s = (0, 0.2, 0.001));
```
